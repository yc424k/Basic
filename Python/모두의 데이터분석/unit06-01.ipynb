```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
april = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '04':
            april.append(float(row[4]))

plt.hist(april, bins = 100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

jan = []
april = []

for row in data:
    month = row[0].split('-')[1]
    if row[4] != '':
        if month == '04':
            april.append(float(row[4]))
        if month == '01':
            jan.append(float(row[4]))

plt.hist(april, bins = 100, color='r', label = 'Apr')
plt.hist(jan, bins = 100, color='b', label='Jan')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

month =[[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data:
    if row[4] != '' :
        month[int(row[0].split('-')[1])-1].append(float(row[4]))

plt.boxplot(month)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)

day = []

for i in range(31):
    day.append([])

for row in data:
    if row[4] != '':
        if row[0].split('-')[1] == '04' :
            day[int(row[0].split('-')[2])-1].append(float(row[4]))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.boxplot(day, showfliers=False)
plt.show()
```
