```python
import csv
import matplotlib.pyplot as plt

f = open('age2.csv', encoding='cp949')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍변동 단위)을 입력해주세요 : ')

for row in data :
    if name in row[0] :
        for i in row[3:104]:
            result.append(int(i.replace(',','')))

print(result)
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.title(name + '지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

for row in data:
    if '탕정면' in row[0]:
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:207]:
            f.append(int(i))

plt.style.use('ggplot')
plt.barh(range(101), m, label='Male')
plt.barh(range(101), f, label='Female')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 입려갛세요 : ')

for row in data:
    if name in row[0] :
        for i in row[3:104]:
            m.append(-int(i))
        for i in row[106:207]:
            f.append(int(i))

plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='Male')
plt.barh(range(101), f, label='Female')
plt.legend()
plt.show()
```
